3. **Data Cleaning**:

```python
df.isnull().sum()
```

```python
# because ['Status', 'Notes'] columns doesn't contain any value
df.drop(['Status', 'Notes'], inplace=True,axis=1)
df.shape
```

```python
df['Benefits'].fillna(0, inplace=True)
```

```python
#for this rows i delet it because its not contatain any value exept of year and agency so it's wrong or uncomplete enter data so it's better to delet it
not_provided_rows = df[df['JobTitle']== 'Not provided']
df.drop(df[df['JobTitle'] == 'Not provided'].index, inplace=True)
not_provided_rows
```

```python
#there are 609 null value in the BasePay i will fill it according to the rule
df['BasePay'].fillna(df['TotalPay'] - df['OvertimePay'] - df['OtherPay'], inplace=True)
```

```python
df.isnull().sum()
```

4. **Basic Data Visualization**: Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.

```python
# I wasn't sure wich column between totalpay or totalbenefit i should use,but the output is the same so i perform only
import matplotlib.pyplot as plt

plt.hist(df['TotalPay'],color='lightblue', edgecolor='black')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.title('Distribution of Total Pay')
plt.show()
```

```python
jobtitle_counts = df['JobTitle'].value_counts().head(10)
custom_colors = ['#ADD8E6', '#FFD700', '#90EE90', '#FFA07A', '#EE82EE', '#FFFF99', '#FFB6C1', '#00FFFF', '#FFC0CB', '#98FB98']

plt.figure(figsize=(8, 8))
plt.pie(jobtitle_counts, labels=jobtitle_counts.index, autopct='%1.1f%%', startangle=140, colors=custom_colors)
plt.title('Employee Distribution by Department')
plt.axis('equal')
plt.show()
```

5. **Grouped Analysis**: Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.

```python
df.groupby('JobTitle')['TotalPay'].describe()
```

```python
df.groupby('JobTitle')['TotalPay'].mean().sort_values(ascending=False)
```

6. **Simple Correlation Analysis**: Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.

```python
salaries_corr = df.corr()['TotalPay']

highest_correlation_column = salaries_corr.drop('TotalPay').idxmax()

plt.figure(figsize=(8, 8))
plt.scatter(df[highest_correlation_column], df['TotalPay'], color='lightblue', alpha=0.5)
plt.title(f'Scatter Plot of TotalPay vs {highest_correlation_column}')
plt.xlabel(highest_correlation_column)
plt.ylabel('TotalPay')
plt.grid(True)
plt.show()
```
